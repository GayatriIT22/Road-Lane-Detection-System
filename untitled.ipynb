```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow # Special patch for Colab
```

```python
def detect_lanes(image_path):
    # Load image
    img = cv2.imread(image_path)

    # Check if image was loaded successfully
    if img is None:
        print(f"Error: Could not load image from {image_path}")
        return None, None

    # 1. Grayscale and Blur
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    blur = cv2.GaussianBlur(gray, (5, 5), 0)

    # 2. Canny Edge Detection
    edges = cv2.Canny(blur, 50, 150)

    # 3. Define a simple triangular mask (Region of Interest)
    height, width = edges.shape
    mask = np.zeros_like(edges)
    # This triangle focuses on the bottom half of the image
    polygon = np.array([[(0, height), (width, height), (width//2, height//2)]], np.int32)
    cv2.fillPoly(mask, polygon, 255)

    # Apply mask
    cropped_edges = cv2.bitwise_and(edges, mask)

    # Show the result in Colab
    print("Original vs Processed Edges")
    cv2_imshow(img)
    cv2_imshow(cropped_edges)

    return img, cropped_edges

# Replace 'your_image.jpg' with the name of the file you uploaded
img, cropped_edges = detect_lanes('/content/Road.PNG')
```

```python
def draw_lines(img, lines):
    # Create a blank image to draw lines on
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)

    if lines is not None:
        for line in lines:
            for x1, y1, x2, y2 in line:
                # Draw lines in green (0, 255, 0) with thickness 5
                cv2.line(line_img, (x1, y1), (x2, y2), (0, 255, 0), thickness=5)

    # Merge the line image with the original image
    combined_img = cv2.addWeighted(img, 0.8, line_img, 1, 1)
    return combined_img

# --- PROCESS ---
# 1. Detect lines from your 'cropped_edges' (the output from your previous step)
# Parameters: rho=2, theta=1 degree, threshold=100, minLineLength=40, maxLineGap=5
lines = cv2.HoughLinesP(cropped_edges, 2, np.pi/180, 100, np.array([]), minLineLength=40, maxLineGap=5)

# 2. Draw them
final_result = draw_lines(img, lines)

# 3. View it
cv2_imshow(final_result)
```

```python
import numpy as np
import cv2

def process_video_frame(frame):
    # 1. Pre-processing
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    blur = cv2.GaussianBlur(gray, (5, 5), 0)
    edges = cv2.Canny(blur, 50, 150)

    # 2. Region of Interest (ROI)
    height, width = edges.shape
    mask = np.zeros_like(edges)
    # Adjust these coordinates based on your specific video perspective
    polygon = np.array([[(0, height), (width, height), (width//2, height//2 + 50)]], np.int32)
    cv2.fillPoly(mask, polygon, 255)
    cropped_edges = cv2.bitwise_and(edges, mask)

    # 3. Hough Transform
    lines = cv2.HoughLinesP(cropped_edges, 2, np.pi/180, 50, np.array([]), minLineLength=40, maxLineGap=100)

    # 4. Draw Lines on a blank canvas
    line_img = np.zeros_like(frame)
    if lines is not None:
        for line in lines:
            x1, y1, x2, y2 = line[0]
            cv2.line(line_img, (x1, y1), (x2, y2), (0, 255, 0), 10)

    # 5. Combine with original frame
    return cv2.addWeighted(frame, 0.8, line_img, 1.0, 0.0)
```
